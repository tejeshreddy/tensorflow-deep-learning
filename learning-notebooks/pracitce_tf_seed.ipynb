```python
import tensorflow as tf
```

```python
t1 = tf.constant([[1, 2], 
                 [3, 4], 
                 [5, 6]])
```

```python
tf.random.shuffle(t1, seed=42)
```

```python
t2 = tf.constant([[11, 12], 
                 [13, 14], 
                 [15, 16]])
```

```python
tf.random.set_seed(42)
tf.random.shuffle(t2)
```
